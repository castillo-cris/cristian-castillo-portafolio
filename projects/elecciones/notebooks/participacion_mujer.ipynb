  <h3>üîß Transformaciones aplicadas</h3>

    <li>üìå Renombrar la columna <code>ano</code> ‚Üí <code>anio</code>.</li>

    <li>üóëÔ∏è Eliminar la columna innecesaria <code>rcd___id</code>.</li>

  <h3>üîß Applied transformations</h3>
  <ul>
    <li>üìå Rename column <code>ano</code> ‚Üí <code>anio</code>.</li>
    <li>üî¢ Convert <code>votos</code> to integer, removing non-numeric characters.</li>
    <li>üóëÔ∏è Drop the unnecessary column <code>rcd___id</code>.</li>
    <li>üè∑Ô∏è Cast <code>genero</code>, <code>departamento</code>, and <code>eleccion</code> to categorical type.</li>
  </ul>
</div>

```python
ruta_dataset = '../data/raw/participacion_mujer.csv'
dataset = pd.read_csv(ruta_dataset, low_memory=False)
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.rename(columns={'ano': 'anio'})

dataset['votos'] = dataset['votos'].str.replace(r'\D', '', regex=True)
dataset['votos'] = dataset['votos'].replace('', '0').astype(int)

dataset = dataset.drop(columns={'rcd___id'})
categorical_cols = ['genero', 'departamento', 'eleccion']
dataset[categorical_cols] = dataset[categorical_cols].astype('category')

print(dataset.dtypes)
```

<div style="background-color:#d9d9d9; padding:15px; border-radius:8px;">
  <h2>üìä An√°lisis de g√©nero en C√°mara y Senado / Gender analysis in C√°mara and Senado</h2>
  <p>En esta secci√≥n se presentan dos perspectivas complementarias sobre la participaci√≥n pol√≠tica en Colombia:</p>
  <ul>
    <li>üó≥Ô∏è <strong>Votos por g√©nero</strong>: evoluci√≥n de los votos recibidos por hombres y mujeres en C√°mara y Senado.</li>
    <li>üë©‚Äçüíº <strong>Candidaturas por g√©nero</strong>: n√∫mero de hombres y mujeres que se postularon en cada corporaci√≥n.</li>
  </ul>
  <p>Ambos gr√°ficos permiten observar tendencias hist√≥ricas, diferencias en representaci√≥n y avances en la participaci√≥n femenina.</p>

  <p>In this section, two complementary perspectives on political participation in Colombia are presented:</p>
  <ul>
    <li>üó≥Ô∏è <strong>Votes by gender</strong>: evolution of votes received by men and women in C√°mara and Senado.</li>
    <li>üë©‚Äçüíº <strong>Candidacies by gender</strong>: number of men and women who ran for office in each chamber.</li>
  </ul>
  <p>Both charts allow us to observe historical trends, differences in representation, and progress in female participation.</p>
</div>

```python
df_filtrado = dataset[
    dataset['genero'].isin(['Masculino', 'Femenino']) &
    dataset['eleccion'].str.lower().str.contains('c√°mara|senado')
]

votos_agrupados = df_filtrado.groupby(['anio', 'genero', 'eleccion'], observed=True)['votos'].sum().reset_index()

votos_agrupados['grupo'] = votos_agrupados['genero'].astype(str) + ' - ' + votos_agrupados['eleccion'].str.capitalize()

votos_pivot = votos_agrupados.pivot(index='anio', columns='grupo', values='votos')

plt.figure(figsize=(12,6))
for columna in votos_pivot.columns:
    if "C√°mara" in columna:
        estilo = '-'
    elif "Senado" in columna:
        estilo = '--'
    else:
        estilo = '-'
    
    plt.plot(votos_pivot.index, votos_pivot[columna], marker='o', linestyle=estilo, label=columna)

plt.title('Votos por g√©nero y tipo de elecci√≥n a lo largo de los a√±os')
plt.xlabel('A√±o')
plt.ylabel('Votos (millones)')
plt.legend(title='Grupo')
plt.grid(True)

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x/1e6:.1f}M'))

plt.tight_layout()
plt.show()
```

```python
candidatos_agrupados = (
    df_filtrado.groupby(['anio', 'genero', 'eleccion'], observed=True)['candidato_a']
    .nunique()
    .reset_index(name='num_candidatos')
)

candidatos_agrupados['grupo'] = (
    candidatos_agrupados['genero'].astype(str) + ' - ' + candidatos_agrupados['eleccion'].str.capitalize()
)

candidatos_pivot = candidatos_agrupados.pivot(index='anio', columns='grupo', values='num_candidatos')

plt.figure(figsize=(12,6))
for columna in candidatos_pivot.columns:
    if "C√°mara" in columna:
        estilo = '-'
    elif "Senado" in columna:
        estilo = '--'
    else:
        estilo = '-'
    
    plt.plot(candidatos_pivot.index, candidatos_pivot[columna], marker='o', linestyle=estilo, label=columna)

plt.title('Cantidad de candidatos por g√©nero y tipo de elecci√≥n a lo largo de los a√±os')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de candidatos')
plt.legend(title='Grupo')
plt.grid(True)
plt.tight_layout()
plt.show()
```
